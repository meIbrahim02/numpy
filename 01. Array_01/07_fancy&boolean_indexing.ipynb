```python
import numpy as np
```

```python
#fancy indexing
arr3 = np.array([1,2,3,4,5,6,7,8])
idx = [1,3,6]
print(arr3[idx])
```

```python
#boolean indexing
arr4 = np.array([1,2,3,4,5,6,7,8])
print(arr4 > 5)
print(arr4 > 3)
print(arr4 % 2 == 0)
```

```python
#boolean indexing
arr4 = np.array([1,2,3,4,5,6,7,8])
print(arr4[arr4 > 5])
print(arr4[arr4 > 3])
print(arr4[arr4 % 2 == 0])
```
