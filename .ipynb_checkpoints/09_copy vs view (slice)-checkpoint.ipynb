```python
# python lists -> copy
# python list creates a copy of the same array and allot different memory to that array
# numpy array -> view
# numpy array views the same array and makes changes on that same array
```

```python
import numpy as np
```

```python
# copy vs view
```

```python
#copy
list = [1,2,3,4,5]
sub_list = list[1:3]
sub_list[0] = 200
print(sub_list)
print(list)
```

```python
# view
arr = np.array([1,2,3,4,5])
sub_arr = arr[1:3]
print(sub_arr)

sub_arr[0] = 200
print(sub_arr)
print(arr)
```

```python
# to create a copy we use ".copy()"
```

```python
arr2 = np.array([1,2,3,4,5])
sub_arr2 = arr2[1:3].copy()
print(sub_arr)

sub_arr2[0] = 200
print(sub_arr2)
print(arr2)
```
