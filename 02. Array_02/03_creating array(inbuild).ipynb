```python
import numpy as np
```

```python
#create
arr1 = np.zeros((2,3)) # prefill
print(arr1, arr1.shape)
```

```python
#we can change the value 
arr2 = np.zeros((2,3), dtype = "int64") 
print(arr2, arr2.shape)
```

```python
# filling with ones
arr3 = np.ones((2,3), dtype = "int64") 
print(arr3, arr3.shape)
```

```python
# full
# in full function we prefill with paticular value
#we can change the value 
arr4 = np.full((3,4), 86) 
print(arr4, arr4.shape)
```

```python
# eye
# with eye function an identitiy matrix will be created
# only single value will be passed
#we can change the value 
arr5 = np.eye(2)
print(arr5, arr5.shape)
```

```python
# range
# range(start, end, step)
#we can change the value 
arr6 = np.arange(1, 10, 2)
print(arr6, arr6.shape)
```

```python
# linspace
# evenly spaced array is created
# linspace(starting value, ending value, no. of elements in array)
arr7 = np.linspace(1, 100, 20)
print(arr7, arr7.shape)
```
