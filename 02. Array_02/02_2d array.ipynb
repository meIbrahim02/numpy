```python
import numpy as np
```

```python
arr = np.array([1, 2, 3, 4, 5, 6, 7])
print(arr, type(arr), arr.dtype, arr.shape)
```

```python
arr2 = np.array([1, 2, 3, 4, 5, 6, 7, "hello"])
print(arr2, type(arr2), arr2.dtype, arr2.shape)
```

```python
#2d array
arr2d = np.array([[1, 2, 3], [4, 5, 6], [2, 4, 6]])
print(arr2d, arr2d.shape)
# shape means dimension of array
# in 2d array the numbers should be homogenous
```
