```python
import numpy as np
import time
```

```python
arr = np.array([1, 2, 3, 4])
print(arr, type(arr))
```

```python
size = 10_000_000

#time
py_list = list(range(size))
start = time.time()
sq_list = [x**2 for x in py_list]
end = time.time()
print(f"python list time is {end - start} seconds")

np_arr = np.array(py_list)
start = time.time()
sq_array = np_arr ** 2
end = time.time()
print(f"numpy array time is {end - start} seconds")
```

```python
#memory
import sys

print(f"python list size of {sys.getsizeof(py_list) * len(py_list)}")

print(f"numpy array size of {np_arr.nbytes} bytes")
```
